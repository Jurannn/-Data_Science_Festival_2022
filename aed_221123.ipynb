# 전체

```python
aed = pd.read_csv(r"C:\Users\USER\Desktop\데사페\AED동대문구.csv", encoding='utf-8')
```

```python
aed.shape
```

```python
ddm_map = folium.Map(location=[37.5744,127.0400], zoom_start=14)
```

```python
aed_loc = aed.loc[:,('위도', '경도')]
```

```python
for i in range(len(aed_loc)):
    folium.Marker([aed_loc.iloc[i][0], aed_loc.iloc[i][1]], popup=aed_loc.index[i]).add_to(ddm_map)
    folium.Circle([aed_loc.iloc[i][0], aed_loc.iloc[i][1]], radius = 50, popup=aed_loc.index[i], color = 'red', fill = True).add_to(ddm_map)
    #folium.Circle([aed_loc.iloc[i][0], aed_loc.iloc[i][1]], radius = 100, popup=aed_loc.index[i], color = 'blue', fill = True).add_to(ddm_map)
    folium.Circle([aed_loc.iloc[i][0], aed_loc.iloc[i][1]], radius = 200, popup=aed_loc.index[i], color = 'blue', fill = True).add_to(ddm_map)
    #folium.Circle([aed_loc.iloc[i][0], aed_loc.iloc[i][1]], radius = 1000, popup=aed_loc.index[i], color = 'black', fill = False).add_to(ddm_map)


ddm_map
```
